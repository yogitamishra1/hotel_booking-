# The prediction with below test data.
##importing test datset


```python
aa = pd.read_csv("/content/test_data_evaluation_part2.csv")
aa
```

```python
bb = aa.drop(['Unnamed: 0', 'ID', 'Nationality'], axis = 1)
bb.head()
```

```python
bb.describe()
```

```python
bb['Age']= bb['Age'].fillna(39)
bb
```

```python
bb.isnull().sum()
```

```python
X=bb.iloc[:,0:6]
y=bb.iloc[:,-1]
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import RidgeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.svm import SVC, LinearSVC
from sklearn import tree
from xgboost import XGBClassifier
from keras.models import Sequential
from keras.layers import Dense, LSTM, Bidirectional, Dropout
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import recall_score, precision_score, f1_score, classification_report
import tensorflow as tf
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=.33, random_state=0)
```

```python
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

## K fold cross validation

```python
kfold = KFold(n_splits=10)
model = LogisticRegression(max_iter=500)

results = cross_val_score(model, X_test, y_test, cv=kfold)
print("-"*40)
print("Accuracy:",results.mean()*100,"%")
print("-"*40)
```

## SVM 

```python
svm = SVC(kernel='rbf')
svm = svm.fit(X_train, y_train)

sns.heatmap(metrics.confusion_matrix(y_pred = svm.predict(X_test), y_true=y_test, labels = [1, 0]),annot = True,fmt="d",cmap="Blues",linewidths=.2) 
plt.xlabel('Predicted classes')
plt.ylabel('Actual classes')
plt.show()
print("-"*50)
print("Accuracy Score for SVM model:", svm.score(X_test, y_test)*100,'%')
print("-"*50)
```

## Decision Tree

```python
dtg = tree.DecisionTreeClassifier(criterion='gini')
kfold = KFold(n_splits=10)
# Creating the model on Training Data
DTree=dtg.fit(X_train,y_train)
prediction=DTree.predict(X_test)
res1 = cross_val_score(dtg, X_test, y_test,cv=kfold)

print(metrics.confusion_matrix(y_test, prediction))
DT_train_accu=dtg.score(X_train, y_train)
print("-"*40)
print('Train Accuracy of the model', round(DT_train_accu,2)*100)
DT_test_accu=dtg.score(X_test, y_test)
print('Test Accuracy of the model', round(DT_test_accu,2)*100)
print('KFold Accuracy:',res1.mean()*100)
print("-"*40)

sns.heatmap(confusion_matrix(y_test,dtg.predict(X_test)), annot= True, cmap='Blues')

dtc_f1=f1_score(y_test,dtg.predict(X_test))*100
plt.title('F1 Score = {}'.format(dtc_f1))
```

##XGBClassifier

```python
xgb =XGBClassifier()
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
xgb_acc = accuracy_score(y_test, y_pred)
results = cross_val_score(xgb, X_test, y_test,cv=kfold)

print("-"*40)
print('Training Accuracy:', xgb.score(X_train,y_train)*100)
print('Test Accuracy:', xgb.score(X_test,y_test)*100)
print('KFold Accuracy:',results.mean()*100)
print("-"*40)
```

##Gaussian NB

```python
nb= GaussianNB()
nb.fit(X_train, y_train)
pred = nb.predict(X_train)
print( "GaussianNB train = ", accuracy_score(y_train, pred)*100)
pred = nb.predict(X_test)
print("GaussianNB test accuracy = ", accuracy_score(y_test, pred)*100)
print('*'*40)
```

## K Neighbors Classifier

```python
kn = KNeighborsClassifier()
kn.fit(X_train, y_train)
pred = kn.predict(X_train)
print( "KNN train = ", accuracy_score(y_train, pred)*100)
pred = kn.predict (X_test)
print("KNN test value = ", accuracy_score(y_test, pred)*100)
```

##LogisticRegression

```python
logreg = LogisticRegression(solver='liblinear')
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
y_pred_prob = logreg.predict_proba(X_test)[:, 1]
log_acc = accuracy_score(y_test, y_pred)
results = cross_val_score(logreg, X_test, y_test,cv=kfold)

print("-"*40)
print('Training Accuracy:', logreg.score(X_train,y_train)*100)
print('Test Accuracy:', logreg.score(X_test,y_test)*100)
print('KFold Accuracy:',results.mean()*100)
print("-"*40)
```

## RidgeClassifier

```python
Ridg = RidgeClassifier()
Ridg.fit(X_train,y_train)
y_pred = Ridg.predict(X_test)
Ridg_acc = accuracy_score(y_test, y_pred)
results = cross_val_score(Ridg, X_test, y_test,cv=kfold)

print("-"*40)
print('Training Accuracy:', Ridg.score(X_train,y_train)*100)
print('Test Accuracy:', Ridg.score(X_test,y_test)*100)
print('KFold Accuracy:',results.mean()*100)
print("-"*40)
```

## neural network

```python
neural_model = tf.keras.Sequential([tf.keras.layers.Dense(8, ),
                                    tf.keras.layers.Dense(1,name = 'Output', activation = 'sigmoid')],
                                    name = 'neural_model_1', )

neural_model.compile(optimizer= tf.keras.optimizers.Adam(learning_rate= 0.1),
                     loss = tf.keras.losses.BinaryCrossentropy(),
                     metrics = ['accuracy'])

history_neural = neural_model.fit(X_train, y_train, epochs= 30)
print('Model Traning complete')
```

```python
neural_model.evaluate(X_test, y_test)
```

###Train Accuracy

```python
y_pred_neural_train=tf.round(neural_model.predict(X_train))
confusion_matrix(y_train, y_pred_neural_train)
print(classification_report(y_train, y_pred_neural_train))
```

### Test Accuracy

```python
y_pred_neural = tf.round(neural_model.predict(X_test))
confusion_matrix(y_test, y_pred_neural)
print(classification_report(y_test, y_pred_neural))
```
